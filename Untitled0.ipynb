```
# loading matplotlib library
import matplotlib.pyplot as plt
```

```
x= [2,9,3,7]
y= [7,15,5,9]

plt.xlabel("Distance")  #x-axis.  
plt.ylabel("Time")  #y-axis.
plt.plot(x,y,label="railway track") #This will draw a line.[label = "" for giving the labels].
plt.scatter(x,y,s=200,label="railway station",marker="*",color='red') #This will mark the points only[s=200/100/20 for size].
plt.bar(x,y,label="height of railway station building") #To draw bar charts.
plt.grid(color='green')  #This will draw a grid.
plt.legend()  #Will show labels in the plot.
plt.show()  #will show the graph.
```

```
#This will be another example of cricket data visualiztion
players=["Smith","Watson","Bailey","Warner","Paine"]
scores=[120,140,180,69,69]

#now plotting it
plt.xlabel('players')
plt.ylabel('scores')
#plt.bar(players,scores,label="2020 score of Australian Cricket Team")
#plt.pie(scores,labels=players)
plt.pie(scores,labels=players,explode=[0.2,0,0,0,0],shadow=True,autopct="%1.1ff%%")
#plt.grid(color='green')
plt.legend()
plt.show()
```
